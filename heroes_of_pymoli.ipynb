* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
players_count = len(purchase_data["SN"].unique())
players_count
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
purchase_data.head()
```

```python
total_unique_items = purchase_data["Item Name"].nunique()
total_unique_items
```

```python
avg_purchase_price = purchase_data["Price"].mean()
avg_purchase_price
```

```python
total_num_purchases = len(purchase_data["Purchase ID"].unique())
total_num_purchases
```

```python
total_revenue = purchase_data["Price"].sum()
total_revenue
```

```python
#Data Frame Built from Variables
purchase_analysis_total_df = pd.DataFrame({
                "Number of Unique Items": [total_unique_items],
                "Average Price": [avg_purchase_price],
                "Number of Purchases": [total_num_purchases],
                "Total Revenue": [total_revenue]
})
purchase_analysis_total_df.loc[:,"Average Price"] = purchase_analysis_total_df["Average Price"].astype(float).map("${:,.2f}".format)
purchase_analysis_total_df.loc[:,"Total Revenue"] = purchase_analysis_total_df["Total Revenue"].astype(float).map("${:,.2f}".format)
purchase_analysis_total_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Simple Count Methods
male_count = len(purchase_data.groupby(['Gender']).get_group(('Male'))["SN"].unique())
male_percent = male_count / players_count
female_count = len(purchase_data.groupby(['Gender']).get_group(('Female'))["SN"].unique())
female_percent = female_count / players_count
other_count = len(purchase_data.groupby(['Gender']).get_group(('Other / Non-Disclosed'))['SN'].unique())
other_percent = other_count / players_count
```

```python
#Data Frame Grouped by Gender, Resets Index to Gender, Divides Each Gender Count by Gender Column Total
gender_df = purchase_data.groupby('Gender')['SN'].nunique().reset_index()
gender_df['Gender Percent'] = (gender_df['SN'] * 100 / gender_df['SN'].sum()).map("{:,.2f}%".format)
gender_df.rename(columns={'SN':'Gender Count'})
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchasing_analysis_gender_df = purchase_data.groupby('Gender').count()
purchasing_analysis_gender_df.drop(['Purchase ID', 'SN', 'Age', 'Item ID', 'Item Name', 'Price'], axis=1, inplace=True)
purchasing_analysis_gender_df["Average Purchase Price"] = (purchase_data.groupby('Gender')['Price'].mean()).astype(float).map("${:,.2f}".format)
purchasing_analysis_gender_df["Total Purchase Value"] = (purchase_data.groupby('Gender')['Price'].sum()).astype(float).map("${:,.2f}".format)
purchasing_analysis_gender_df["Average Total Purchase per User"] = (purchase_data.groupby('Gender')['Price'].sum() / purchase_data.groupby('Gender')['SN'].nunique()).astype(float).map("${:,.2f}".format)
purchasing_analysis_gender_df.rename(columns={'Age Group': 'Purchase Count'})
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 99]
bin_names = ["< 10 Years", "10-14 Years", "15-19 Years", "20-24 Years", "25-29 Years", "30-34 Years", "34-39 Years", "40+ Years"]
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels=bin_names)
purchase_data
```

```python
age_demos_df = purchase_data.groupby('Age Group')['SN'].nunique().reset_index()
age_demos_df["Percent of Age Group"] = (age_demos_df['SN'] * 100 / age_demos_df['SN'].sum()).map("{:,.2f}%".format)
age_demos_df.rename(columns={'SN': 'Age Group Count'})
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_purchasing_analysis_df = purchase_data.groupby('Age Group').count()
age_purchasing_analysis_df['Purchase Count'] = purchase_data.groupby('Age Group')['Purchase ID'].count()
age_purchasing_analysis_df['Average Purchase Price'] = (purchase_data.groupby('Age Group')['Price'].mean()).astype(float).map("${:,.2f}".format)
age_purchasing_analysis_df['Total Purchase Price'] = (purchase_data.groupby('Age Group')['Price'].sum()).astype(float).map("${:,.2f}".format)
age_purchasing_analysis_df['Purchase Total Per User'] = round(purchase_data.groupby('Age Group')['Price'].mean(), 2)
age_purchasing_analysis_df.drop(['Purchase ID', 'SN', 'Age', 'Gender', 'Item ID', 'Item Name', 'Price'], axis=1, inplace=True)
age_purchasing_analysis_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_spenders_df = purchase_data.groupby('SN').count()
top_spenders_df.drop(['Age Group', 'Age', 'Gender', 'Item ID', 'Item Name', 'Price'], axis=1, inplace=True)
top_spenders_df['Average Purchase Price'] = (purchase_data.groupby('SN')['Price'].mean()).astype(float).map("${:,.2f}".format)
top_spenders_df['Total Purchase Price'] = (purchase_data.groupby('SN')['Price'].sum()).astype(float).map("${:,.2f}".format)
top_spenders_df.sort_values('Purchase ID', ascending=False)
top_spenders_df.rename(columns={'Purchase ID': 'Purchase Count'})
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
pop_items_df = purchase_data.groupby(['Item ID', 'Item Name','Price'])['Price'].agg(['count','sum'])
pop_items_df.columns = ['Purchase Count', 'Total Purchase Value']
most_pop_items_df = pop_items_df.sort_values(by='Purchase Count', ascending=False)
most_pop_items_df['Total Purchase Value'] = most_pop_items_df['Total Purchase Value'].map('${:,.2f}'.format)
most_pop_items_df
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
most_pop_items_df = pop_items_df.sort_values(by='Total Purchase Value', ascending=False)
most_pop_items_df['Total Purchase Value'] = most_pop_items_df['Total Purchase Value'].map('${:,.2f}'.format)
most_pop_items_df
```
