```python
import pandas as pd
import numpy as np
```

```python
file = "Resources/purchase_data.csv"
```

```python
purchase_data = pd.read_csv(file)
```

```python
purchase_data.head() #table showing columns of purchasing data
```

```python
purchase_data["SN"].unique() #array of unique player names
```

```python
unique_players = purchase_data["SN"].value_counts() #total unique players = 576
unique_players
```

```python
unique_items = purchase_data["Item Name"].value_counts() #total of unique items = 179
unique_items
```

```python
total_purchases = unique_items.sum() #total purchases = 780
total_purchases
```

```python
average_price = round(purchase_data["Price"].mean(), 2)
average_price
```

```python
total_revenue = purchase_data["Price"].sum()
total_revenue
```

```python
u_items_count = unique_items.count()
u_items_count
```

```python
purchase_summary_df = pd.DataFrame({
                "Number of Unique Items": [u_items_count],
                "Average Price": [average_price],
                "Number of Purchases": [total_purchases],
                "Total Revenue": [total_revenue]
})
purchase_summary_df.loc[:,"Average Price"] = purchase_summary_df["Average Price"].astype(float).map("${:,.2f}".format)
purchase_summary_df.loc[:,"Total Revenue"] = purchase_summary_df["Total Revenue"].astype(float).map("${:,.2f}".format)
purchase_summary_df
```
